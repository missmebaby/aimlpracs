```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the heart disease dataset
df = pd.read_csv('/content/diabetes_prediction_dataset.csv')

# Display the first few rows of the dataset
df.head(5)

```

```python
# Get the summary statistics of the dataset
print(df.describe())
```

```python
# Check for missing values
print(df.isnull().sum())
```

```python
# Remove Unneccessary value [0.00195r
df = df[df['gender'] != 'Other']
```

```python
# Visualize the distribution of gender
plt.figure(figsize=(8, 6))
sns.countplot(x='gender', data=df)
plt.title('Gender Distribution')

plt.show()
```

```python
# Visualize the correlation between variables using a heatmap
plt.figure(figsize=(10, 8))
sns.heatmap(df.corr(), annot=True, cmap="coolwarm")
plt.title('Correlation Heatmap')
plt.show()
```

```python
# Visualize the distribution of bmi with respect to density
plt.figure(figsize=(10, 6))
sns.distplot(df['bmi'],  kde=True)
plt.title('BMI Distribution')
plt.show()
```

```python
# Visualize the distribution of bmi among patients with and without diabetes
sns.boxplot(x='diabetes', y='bmi', data=df)
plt.title('BMI vs Diabetes')
plt.show()
```

```python
# Visualize the relationship between age and bmi
sns.scatterplot(x='age', y='bmi', hue='diabetes', data=df)
plt.title('Age vs BMI')
plt.show()
```
